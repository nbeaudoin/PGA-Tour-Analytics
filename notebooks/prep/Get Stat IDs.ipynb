```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re
import os.path
from os import path
from urllib.parse import urlparse
```

## Create Stat IDs from page for each main and sub category

```python
stat_id = '001'
year = '2020'

url = "https://www.pgatour.com/stats/categories.ROTT_INQ.html"
html = urlopen(url)
soup = BeautifulSoup(html)

url_list = []
for link in soup.findAll('a', attrs={'href': re.compile("^/stats/stat.")}):
    #print(link.get('href'))
    url_list.append(link.get('href'))
    
url_list[:10]
```

```python
# Create list of stat_ids
stat_id_list = []
for url in url_list:
    new_url = url[12:]
    stat_id = new_url[:-5]
    stat_id_list.append(stat_id)
    
stat_id_list[:10]
```

```python
# Iterate over all pages to get Stat IDs

categories = ['ROTT', 'RAPP', 'RARG', 'RPUT', 'RSCR', 'RSTR']

stat_id_list = []

for category in categories:
    
    url = 'https://www.pgatour.com/stats/categories.{}_INQ.html'.format(category)
    html = urlopen(url)
    soup = BeautifulSoup(html)

    url_list = []
    for link in soup.findAll('a', attrs={'href': re.compile("^/stats/stat.")}):
        #print(link.get('href'))
        url_list.append(link.get('href'))
    
    for url in url_list:
        new_url = url[12:]
        stat_id = new_url[:-5]
        stat_id_list.append(stat_id)
    
    print(category, "complete")
```

```python
# Create dataframe for later use
# dictionary of lists  
dict = {'Stat_ID': stat_id_list}  
    
df = pd.DataFrame.from_dict(dict, orient='index').T
```
